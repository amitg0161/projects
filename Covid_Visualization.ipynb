```python
import covid
import matplotlib.pyplot as plt

cov = covid.Covid()
```

```python
#check the list of the counties and their codes
cov.list_countries()
```

```python
#taking input for the country name

name = input("enter the country name: ")
virusdata = cov.get_status_by_country_name(name)
```

```python
#getting the cases

active = virusdata['active']
recover = virusdata['recovered']
deaths = virusdata['deaths']

val = [active, recover, deaths]
```

```python
#visualizing the cases

plt.figure(figsize=(8, 8))
plt.pie(val, labels=['Active', 'Recovered', 'Deaths'], autopct=lambda p : '{:.2f}%  ({:,.0f})'.format(p,p * sum(val)/100))
plt.title("Visualiztion of Covid cases in " + name)
plt.legend(loc = 'lower left')
plt.show()
```
